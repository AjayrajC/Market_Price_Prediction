```python
import pandas as pd

# Load the dataset from CSV file
df = pd.read_csv('/content/MarketPricePrediction.csv')

# Function to find distinct values in each column
def find_distinct_values(dataframe):
    distinct_values = {}
    for column in dataframe.columns:
        distinct_values[column] = dataframe[column].unique()
    return distinct_values

# Function to convert categorical values to numeric
def convert_categorical_to_numeric(dataframe):
    for column in dataframe.columns:
        if dataframe[column].dtype == 'object':
            # Check if the column can be converted to datetime with just month and year
            try:
                dataframe[column] = pd.to_datetime(dataframe[column], format='%B-%Y')
                # Extract year and month as separate columns
                dataframe[column + '_year'] = dataframe[column].dt.year
                dataframe[column + '_month'] = dataframe[column].dt.month
                # Drop the original date column
                dataframe.drop(column, axis=1, inplace=True)
            except (ValueError, TypeError):
                # If not a date, convert to categorical codes
                dataframe[column] = dataframe[column].astype('category').cat.codes
    return dataframe

# Step 1: Find distinct values in each column
distinct_values = find_distinct_values(df)
print("Distinct values in each column:")
for col, values in distinct_values.items():
    print(f"{col}: {values}")

# Step 2: Convert categorical values to numeric, handling dates separately
df_encoded = convert_categorical_to_numeric(df)
print("\nDataFrame after categorical conversion:")
print(df_encoded.head())

# Save the encoded DataFrame to a new CSV file
df_encoded.to_csv('new_market_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the modified dataset
df = pd.read_csv('/content/modified_market_data.csv')

# Step 1: Inspect the Data
print(df.info())
print(df.describe())

# Step 2: Temporal Analysis
# Combine year and month into a single datetime column (setting day to 1 by default)
df['date'] = pd.to_datetime(df['date_year'].astype(str) + '-' + df['date_month'].astype(str) + '-01')

print(df.head())

# Plot quantity over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='quantity')
plt.title('Quantity Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()

# Plot priceMod over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='priceMod')
plt.title('Modal Price Over Time')
plt.xlabel('Date')
plt.ylabel('Modal Price')
plt.grid(True)
plt.show()

# Step 3: Seasonality and Trends
# Group by year and month to find average trends
monthly_avg = df.groupby(['date_year', 'date_month']).mean().reset_index()
monthly_avg['date'] = pd.to_datetime(monthly_avg['date_year'].astype(str) + '-' + monthly_avg['date_month'].astype(str) + '-01')

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_avg, x='date', y='quantity')
plt.title('Monthly Average Quantity')
plt.xlabel('Date')
plt.ylabel('Average Quantity')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_avg, x='date', y='priceMod')
plt.title('Monthly Average Modal Price')
plt.xlabel('Date')
plt.ylabel('Average Modal Price')
plt.grid(True)
plt.show()

# Step 4: Detect Anomalies
# Using IQR to detect anomalies in quantity
Q1 = df['quantity'].quantile(0.25)
Q3 = df['quantity'].quantile(0.75)
IQR = Q3 - Q1
anomalies_quantity = df[(df['quantity'] < (Q1 - 1.5 * IQR)) | (df['quantity'] > (Q3 + 1.5 * IQR))]

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='date', y='quantity', label='Normal')
sns.scatterplot(data=anomalies_quantity, x='date', y='quantity', color='red', label='Anomaly')
plt.title('Quantity Anomalies Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.grid(True)
plt.show()

# Using IQR to detect anomalies in priceMod
Q1 = df['priceMod'].quantile(0.25)
Q3 = df['priceMod'].quantile(0.75)
IQR = Q3 - Q1
anomalies_priceMod = df[(df['priceMod'] < (Q1 - 1.5 * IQR)) | (df['priceMod'] > (Q3 + 1.5 * IQR))]

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='date', y='priceMod', label='Normal')
sns.scatterplot(data=anomalies_priceMod, x='date', y='priceMod', color='red', label='Anomaly')
plt.title('Modal Price Anomalies Over Time')
plt.xlabel('Date')
plt.ylabel('Modal Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the modified dataset
df = pd.read_csv('/content/modified_market_data.csv', parse_dates=['date'])

# Sort the DataFrame by date
df.sort_values(by='date', inplace=True)

# Step 1: Create Lagged Variables
# Creating lag features for 'quantity' and 'priceMod'
df['quantity_lag1'] = df['quantity'].shift(1)
df['quantity_lag2'] = df['quantity'].shift(2)
df['priceMod_lag1'] = df['priceMod'].shift(1)
df['priceMod_lag2'] = df['priceMod'].shift(2)

# Step 2: Create Rolling Statistics
# Creating rolling mean and standard deviation for 'quantity' and 'priceMod'
df['quantity_roll_mean3'] = df['quantity'].rolling(window=3).mean()
df['quantity_roll_std3'] = df['quantity'].rolling(window=3).std()
df['priceMod_roll_mean3'] = df['priceMod'].rolling(window=3).mean()
df['priceMod_roll_std3'] = df['priceMod'].rolling(window=3).std()

# Step 3: Create Seasonal Indicators
# Extracting month and year as separate features
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# Adding seasonal dummy variables (for each month)
df = pd.get_dummies(df, columns=['month'], drop_first=True)

# Drop rows with NaN values generated due to lag and rolling operations
df.dropna(inplace=True)

# Display the DataFrame to verify the new features
print(df.head())

# Save the DataFrame with new features to a new CSV file
df.to_csv('feature_engineered_market_data.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load the feature-engineered dataset
df = pd.read_csv('/content/feature_engineered_market_data.csv')

# Splitting the data into train and test sets (80% train, 20% test)
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Define evaluation function
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    return mae, mse, rmse

# ARIMA Model
def arima_model(train, test):
    model = ARIMA(train['quantity'], order=(5,1,0))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=len(test))
    mae, mse, rmse = evaluate_model(test['quantity'], forecast)
    return mae, mse, rmse

# SARIMA Model
def sarima_model(train, test):
    model = SARIMAX(train['quantity'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=len(test))
    mae, mse, rmse = evaluate_model(test['quantity'], forecast)
    return mae, mse, rmse

# LSTM Model
def lstm_model(train, test):
    scaler = MinMaxScaler()
    scaler.fit(train[['quantity']])
    train_scaled = scaler.transform(train[['quantity']])
    test_scaled = scaler.transform(test[['quantity']])

    def create_dataset(X, y, time_steps=1):
        Xs, ys = [], []
        for i in range(len(X) - time_steps):
            v = X[i:(i + time_steps)]
            Xs.append(v)
            ys.append(y[i + time_steps])
        return np.array(Xs), np.array(ys)

    time_steps = 30
    X_train, y_train = create_dataset(train_scaled, train_scaled, time_steps)
    X_test, y_test = create_dataset(test_scaled, test_scaled, time_steps)

    model = Sequential()
    model.add(LSTM(units=64, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.1, verbose=0)

    y_pred = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred)
    y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
    mae, mse, rmse = evaluate_model(y_test, y_pred)
    return mae, mse, rmse

# Evaluate models
arima_mae, arima_mse, arima_rmse = arima_model(train, test)
sarima_mae, sarima_mse, sarima_rmse = sarima_model(train, test)
lstm_mae, lstm_mse, lstm_rmse = lstm_model(train, test)

# Print evaluation results
print("ARIMA Model:")
print(f"MAE: {arima_mae}, MSE: {arima_mse}, RMSE: {arima_rmse}")

print("\nSARIMA Model:")
print(f"MAE: {sarima_mae}, MSE: {sarima_mse}, RMSE: {sarima_rmse}")

print("\nLSTM Model:")
print(f"MAE: {lstm_mae}, MSE: {lstm_mse}, RMSE: {lstm_rmse}")
```

```python
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit, train_test_split
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from itertools import product

# Load the feature-engineered dataset
df = pd.read_csv('/content/feature_engineered_market_data.csv')

# Splitting the data into train and test sets (80% train, 20% test)
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Further split the training data into training and validation sets (80% train, 20% validation)
train_data, val_data = train_test_split(train, test_size=0.2, shuffle=False)

# Define evaluation function
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    return mae, mse, rmse

# Cross-validation function for time series
def time_series_cross_val(data, model_func, param_grid):
    tscv = TimeSeriesSplit(n_splits=5)
    best_score, best_params = float("inf"), None
    for params in param_grid:
        scores = []
        for train_index, val_index in tscv.split(data):
            train_fold, val_fold = data.iloc[train_index], data.iloc[val_index]
            score = model_func(train_fold, val_fold, *params)
            scores.append(score[2])  # Using RMSE for comparison
        avg_score = np.mean(scores)
        if avg_score < best_score:
            best_score, best_params = avg_score, params
    return best_score, best_params

# ARIMA Model
def arima_model(train, test, p, d, q):
    model = ARIMA(train['quantity'], order=(p,d,q))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=len(test))
    return evaluate_model(test['quantity'], forecast)

# SARIMA Model
def sarima_model(train, test, p, d, q, P, D, Q, s):
    model = SARIMAX(train['quantity'], order=(p, d, q), seasonal_order=(P, D, Q, s))
    model_fit = model.fit(disp=False)
    forecast = model_fit.forecast(steps=len(test))
    return evaluate_model(test['quantity'], forecast)

# LSTM Model
def lstm_model(train, test, units, epochs, batch_size):
    scaler = MinMaxScaler()
    scaler.fit(train[['quantity']])
    train_scaled = scaler.transform(train[['quantity']])
    test_scaled = scaler.transform(test[['quantity']])

    def create_dataset(X, time_steps=1):
        Xs, ys = [], []
        for i in range(len(X) - time_steps):
            Xs.append(X[i:(i + time_steps)])
            ys.append(X[i + time_steps])
        return np.array(Xs), np.array(ys)

    time_steps = 30
    X_train, y_train = create_dataset(train_scaled, time_steps)
    X_test, y_test = create_dataset(test_scaled, time_steps)

    model = Sequential()
    model.add(LSTM(units=units, return_sequences=True, input_shape=(time_steps, 1)))
    model.add(LSTM(units=units // 2))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=0)

    y_pred = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred)
    y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
    return evaluate_model(y_test, y_pred)

# Parameter grids
arima_param_grid = list(product(range(3), range(2), range(3)))  # Simple grid for (p,d,q)
sarima_param_grid = list(product(range(2), range(2), range(2), range(2), range(2), range(2), [12]))  # (p,d,q,P,D,Q,s)
lstm_param_grid = list(product([32, 64, 128], [50, 100], [16, 32]))  # (units, epochs, batch_size)

# Evaluate models with cross-validation
best_arima_score, best_arima_params = time_series_cross_val(train, arima_model, arima_param_grid)
best_sarima_score, best_sarima_params = time_series_cross_val(train, sarima_model, sarima_param_grid)
best_lstm_score, best_lstm_params = time_series_cross_val(train, lstm_model, lstm_param_grid)

# Final model evaluation on test set
arima_mae, arima_mse, arima_rmse = arima_model(train, test, *best_arima_params)
sarima_mae, sarima_mse, sarima_rmse = sarima_model(train, test, *best_sarima_params)
lstm_mae, lstm_mse, lstm_rmse = lstm_model(train, test, *best_lstm_params)

# Print evaluation results
print("ARIMA Model:")
print(f"Best Params: {best_arima_params}")
print(f"MAE: {arima_mae}, MSE: {arima_mse}, RMSE: {arima_rmse}")

print("\nSARIMA Model:")
print(f"Best Params: {best_sarima_params}")
print(f"MAE: {sarima_mae}, MSE: {sarima_mse}, RMSE: {sarima_rmse}")

print("\nLSTM Model:")
print(f"Best Params: {best_lstm_params}")
print(f"MAE: {lstm_mae}, MSE: {lstm_mse}, RMSE: {lstm_rmse}")
```
